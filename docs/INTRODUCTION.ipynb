## Azure OpenAI
>Azure Open AI是一种人工智能服务，用户可以通过简单的API调用实现多种AI技术。该服务可以帮助开发者构建应用程序，以从图像、语音中提取有用信息，理解方言和情感，以及生成先进的AI模型。

    服务模型
    - GPT-4 Turbo
    - GPT-4
    - GPT-3.5
    - DALL-E
    - Whisper
    - Text to speech

## Open WebUI
> Open WebUI是为了实现完全离线操作而设计的可扩展、功能丰富且用户友好的自托管WebUI。它支持多种LLM runner，包括Ollama和OpenAI兼容的API。

Open WebUI有以下优点：
 - 直观的界面
 - 易于安装
 - 多模型对话
 - 良好的用户管理

## 应用安装

```
# Node.js >= 20.10
# Python >= 3.11

git clone -b virtual-human-dev https://github.com/zqgu2016/open-webui.git
cd open-webui/

# Copying required .env file
cp -RPp .env.example .env

# Building Frontend Using Node
npm i
npm run build

# Serving Frontend with the Backend
cd ./backend
pip install -r requirements.txt -U
bash start.sh 
```
